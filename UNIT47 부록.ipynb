## 47.8 with as에 사용할 수 있는 클래스 만들기

#### open으로 파일을 열때 with as 사용 파일 객체의 close 자동 호출
#### with as 사용하려면 클래스에 __enter__ 와 __exit__ 메서드 구현

````python
class OpenHello:
    
    def __init__(self, filename):
        import os
        os.mkdir('code_UNIT47')
        self.path = 'code_UNIT47/'
        self.filename = filename
    
    def __enter__(self):
        self.file = open(self.path + self.filename, 'w') # 파일 객체를 속성에 저장
        return self.file # __enter__에서 값을 반환하면 as에 지정한 변수에 들어감
    
    def __exit__(self, exc_type, exc_value, trackback):
        self.file.close() # __exit__에서 파일 객체 닫기
        
with OpenHello('hello.text') as hello:
    hello.write('Hello, world')
````

````python
with open('code_UNIT47/hello.text', 'r') as file:
    r = file.read()
    print(r)
````

````python
# list 상속받고, 속성 desc와 메서드 replace가 들어있는 클래스 만들기

def replace(self, old, new):
    while old in self:
        self[self.index(old)] = new
        
# list를 상속받음, 속성  desc, 메서드 replace 추가
AdvanceList = type('AdvanceList', (list, ),
                  {'desc': '향상된 리스트', 'replace':replace})

x = AdvanceList([1,2,3,1,2,3,1,2,3])
x
````

### 47.9.2 type을 상속받아서 메타클래스 구현하기

#### 메타클래스 __new__
    클래스가 type을 상속받으면 메타클래스가 됨
    __new__ 메서드에서 새로 만들어질 클래스에 속성과 메서드를 추가해 줄 수 있음

````python
class MakeCalc(type): # type을 상속받으므로 메타클래스가 됨
    def __new__(metacls, name, bases, namespace): # 새 클래스를 만들 때 호출되는 메서드
        namespace['desc'] = '계산 클래스'
        namespace['add'] = lambda self, a, b: a+b # 새 클래스에 메서드 추가
        return type.__new__(metacls, name, bases, namespace) # type의 __new__호출
    
Clac = MakeCalc('Calc', (), {}) # 메타클래스 MakeCalc로 클래스 Calc 생성
c = Clac() # 클래스 Calc로 인스턴스 c 생성

print(c.desc)
print(c.add(1, 2))
````

### 47.9.3 메타클래스 활용하기

#### 메타클래스는 주로 클래스의 동작을 제어할 때 사용
#### 싱글톤(Singleton)방식 
#### 클래스의 인스턴스를 언제나 하나만 생성

````python
# singleton.py
class Singleton(type): # type을 상속받음 <- 메타클래스 
    __instances = {} # 클래스의 인스턴스를 저장할 속성
    def __call__(cls, *args, **kwargs): # 클래스로 인스턴스를 만들 때 호출되는 메서드
        if cls not in cls.__instances: # 클래스로 인스턴스를 생성하지 않았는지 확인
            cls.__instances[cls] = super().__call__(*args, **kwargs)
            # 생성하지 않았으면 인스턴스를 생성하여 속성에 저장
        return cls.__instances[cls] # 클래스로 인스턴스를 생성했으면 인스턴스 반환
    
    
class Hello(metaclass=Singleton): # 메타클래스로 Singleton을 지정
    pass

a = Hello()
b = Hello()
print(a is b)
````

### 47.10.2 await로 네이티브 코루틴 실행하기

#### await로 네이티브 코루틴을 실행하는 방법
#### await뒤에 코루틴 객체, 퓨처 객체, 태스크 객체를 지정
#### 해당 객체가 끝날 때까지 기다린 뒤 결과를 반환
#### await는 특정 객체가 끝날 때까지 기다림

    변수 = await 코루틴객체
    변수 = await 퓨처객체
    변수 = await 태스크객체

```
import asyncio

async def add(a, b):
    print(f'add : {a} + {b}')
    await asyncio.sleep(1.0) # 1초 대기, asyncio.sleep도 네이티브 코루틴
    return a + b

async def print_add(a, b):
    result = await add(a, b) # await로 다른 네이티브 코루틴 실행
                             # 반환값을 변수에 저장
    print(f'print_add : {a} + {b} = {result}')

loop = asyncio.get_event_loop() # 이벤트 루프를 얻음
loop.run_until_complete(print_add(1, 2)) # print_add가 끝날 때까지 이벤트 루프를 실행
loop.close() # 이벤트 루프를 닫음
```

#### 퓨처와 태스크
    퓨처(asyncio.Future)는 미래에 할 일을 표현하는 클래스
        할일을 취소하거나 상태 확인, 완료 및 결과 설정에 사용
    태스크(asyncio.Task)는 퓨처의 파생 클래스
        퓨처의 기능과 실행할 코루틴의 객체를 포함
        태스크는 코루틴 실행을 취소하거나 상태 확인, 완료 및 결과 설정에 사용

### 47.10.4 웹 페이지를 비동기로 가져오기

```
from time import time
from urllib.request import Request, urlopen
import asyncio

urls = ['https://www.google.co.kr/search?q=' + i for i in ['apple','pear','grape','pineapple', 'orange', 'strawberry']]



async def fetch(url):
    request = Request(url, headers={'User-Agent':'Mozilla/5.0'}) # UA가 없으면 403 에러 발생
    
    response = await loop.run_in_executor(None, urlopen, request) # run_in_executor 사용

    page = await loop.run_in_executor(None, response.read) # run_in_executor 사용

    return len(page)


async def main():
    futures = [asyncio.ensure_future(fetch(url)) for url in urls] # 태스크(퓨처) 객체를 리스트로 만듦
                                                                  
    result = await asyncio.gather(*futures) # 결과를 한꺼번에 가져옴
    print(result)


begin = time()
loop = asyncio.get_event_loop() # 이벤트 루프를 얻음
loop.run_until_complete(main()) # main이 끝날 때까지 기다림
loop.close()
end = time()
print(f'실행시간 : {end - begin:.3f}')

```

#### urlopen이나 response.read 같은 메서드는 결과가 나올 때 까지 코드 실행 중단
#### 이런 함수들을 블로킹 I/O 라고 부름
#### 네이티브 코루틴 안에서 블로킹 함수를 실행하려면
#### 이벤트 루프의 run_in_executor함수를 사용
#### 다른 스레드에서 병렬로 실행시켜야 함

#### run_in_executor 첫 번째 인수는 executor 
#### 함수를 실행 시켜줄 스레드 풀 또는 프로세스 풀
#### None을 넣어서 기본 스레드 풀을 사용
#### 두 번째 인수에는 실행할 함수를 넣음
#### 세 번째 인수부터는 인수를 차례대로 넣음

### 47.10.6 async with

        async def __aexit__(self, exc_type, exc_value, traceback):
            코드
            

````python
# 1초 뒤에 덧셈 결과를 반환하는 클래스

import asyncio

class AsyncAdd:
    def __init__(self, a, b):
        self.a = a
        self.b = b
        
    async def __aenter__(self):
        await asyncio.sleep(1.0)
        return self.a + self.b # __aenter__에서 값을 반환하면 as에 지정한 변수에 들어감
    
    async def __aexit__(self, exc_type, exc_value, traceback):
        pass
    
async def main():
    async with AsyncAdd(1, 2) as result: # async with에 클래스의 인스턴스 지정
        print(result)
        
loop = asyncio.get_event_loop()
loop.run_until_complete(main())
loop.close()
````

### 47.10.7 async for

#### __aiter__와 __anext__ 메서드 구현해야 함 

````python
# async_for.py
import asyncio

class AsyncCounter:
    def __init__(self, stop):
        self.current = 0
        self.stop = stop
        
    def __aiter__(self):
        return self
    
    async def __anext__(self):
        if self.current < self.stop:
            await asyncio.sleep(1.0)
            r = self.current
            self.current += 1
            return r
        else:
            raise StopAsyncIteration
        
async def main():
    async for i in AsyncCounter(3):
        print(i, end=' ')
        
loop = asyncio.get_event_loop()
loop.run_until_complete(main())
loop.close()
````

## 47.13 프로퍼티 사용하기

#### 클래스에서 메서드를 통하여 속성의 값을 가져오거나 저장하는 경우가 있음
#### 이때 값을 가져오는 메서드를 getter, 값을 저장하는 메서드를 setter라고 부름

````python
# getter_setter.py
class Person:
    def __init__(self):
        self.__age = 0 
    
    # getter
    def get_age(self):
        return self.__age
    
    # setter
    def set_age(self, value):
        self.__age = value
        
james = Person()
james.set_age(20)
print(james.get_age())
````

#### @property를 사용 getter, setter 구현
#### @property
#### @메서드이름.setter

````python
class Person:
    def __init__(self):
        self.__age = 0 
    
    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self, value):
        self.__age = value
        
james = Person()
james.age = 20
print(james.age)
````
