## 31.1 재귀호출 사용하기

```python
def hello():
    print('hello')
    hello()
    
hello()
```

#### 스택넘침 현상 발생함 잘못된 재귀호출은 사용하지 않는것이 좋음

### 31.1.1 재귀호출에 종료 조건 만들기

```python
def hello(count):
    if count == 0:
        return
    print('hello')
    count -= 1
    hello(count)
    
hello(5)
```

## 31.2 재귀호출로 팩토리얼 구하기

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))
```
