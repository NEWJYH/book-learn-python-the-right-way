### 45.1.3 모듈에 클래스 작성하기

````python
# person.py
class Person:
    def __init__(self, name, age, address):
        self.name = name
        self.age = age
        self.address = address

    def greeting(self):
        print(f'안녕하세요. 저는 {self.name}입니다.')
````

````python
import code_UNIT45.person as person

maria = person.Person('마리아', 20, '서울시 서초구 반포자이')
maria.greeting()
````

## 45.2 모듈과 시작점 알아보기

```
    if __name__ == "__main__":
        코드
```

#### 현재 스크립트 파일이 실행되는 상태를 파악하기 위해 사용
#### 파이썬 인터프리터가 최초 실행한 스크립트 파일의 __name__에는 __main__이 들어감
#### 프로그램 시작점(entry point)라는 뜻

````python
# calc.py

def add(a, b):
    return a + b

def mul(a, b):
    return a * b

if __name__ == '__main__': # 프로그램의 시작점일 때만 아래 코드 실행
    print(add(10, 20))
    print(mul(10, 20))
````

````python
import code_UNIT45.calc

code_UNIT45.calc.add(50, 60)
code_UNIT45.calc.mul(50, 60)
````
